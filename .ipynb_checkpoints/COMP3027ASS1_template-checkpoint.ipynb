```python
# This function should prepare the data by reading it from a file and converting it into a useful format for training and testing
# and implement 90-10 splitting as specified in the project description.
import csv
import numpy as np

def preprocess(filename):
    
    # Reads in the dataset
    dataset = list()
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)
        for row in csv_reader:
            dataset.append(row)
    
    #Splits the dataset into a train and a test set
    # 90% of the dataset will be in the training set
    # 10% of the dataset will be in the test set
    train = list()
    test = list(dataset)
    
    split = 0.9
    train_size = split * len(dataset)
    
    while len(train) < train_size:
        train.append(test.pop(0))

    return train, test

```

```python
def calculateIntersection(train_set, attribute_index):
    '''
    Inputs are the training set and the index of the attributee in question.
    Calculates the probability of the attribute intersection above 50k and below 50k.
    Stores it in a dictionary and returns the dictionary.
    '''
    label_index = 11
    
    # Dictionary contains {label:[Pr(A∩"<=50K"),Pr(A∩">50K")]} 
    attribute_dictionary = {}
    
    # Calculate the count for the attribute given the label
    for row in train_set:
        if row[attribute_index] not in attribute_dictionary:
            attribute_dictionary[row[attribute_index]] = [0, 0]
        
        if row[label_index].strip() == "<=50K":
            attribute_dictionary[row[attribute_index]][0] += 1
        elif row[label_index].strip() == ">50K":
            attribute_dictionary[row[attribute_index]][1] += 1
    
    
    # Perform Laplace Smoothing
    for data in attribute_dictionary:
        attribute_dictionary[data][0] += 1
        attribute_dictionary[data][1] += 1
    
    
    # Calculates the intersection
    for data in attribute_dictionary:
        attribute_dictionary[data][0] = (attribute_dictionary[data][0])/len(train_set)
        attribute_dictionary[data][1] = (attribute_dictionary[data][1])/len(train_set)
    
    return attribute_dictionary

def calculateConditionalProbability(dictionary, prior_over50, prior_under50):
    for data in dictionary:
        cond_prob_under50 = dictionary[data][0]/prior_under50
        cond_prob_over50 = dictionary[data][1]/prior_over50
        
        dictionary[data] = [cond_prob_under50, cond_prob_over50]

    return dictionary
```

```python
# This function should calculat prior probabilities and likelihoods (conditional probabilities) from the training data and using
# to build a naive Bayes model

def train(train_set):
    
    label_index = 11
    
    label_list = list()
    age_list_under50 = list()
    age_list_over50 = list()

    hours_list_over50 = list()
    hours_list_under50 = list()

    
    scores = list()
    
    # Contains {label: [Pr(A∩"<=50K"),Pr(A∩">50K")]}
    intersection_dictionary = {}
    
    # Contains {label: [Conditional probability of "<=50K" , Cond prob of ">50K"]}
    conditional_dict = {}
    
    prior_over50 = 0
    prior_under50 = 0
    under50_count = 0
    over50_count = 0
    
    # Calculates the prior probability
    for entry in train_set:
        if entry[label_index].strip() == "<=50K":
            under50_count += 1
        else:
            over50_count += 1
    
    prior_over50 = over50_count/len(train_set)
    prior_under50 = under50_count/len(train_set)    
        
    # Adds each attribute into their own list
    for row in train_set:
        if row[label_index].strip() == "<=50K":
            age_list_under50.append(int(row[0]))
            hours_list_under50.append(int(row[9]))
        else:
            age_list_over50.append(int(row[0]))
            hours_list_over50.append(int(row[9]))


    for i in range(len(train_set[0])-1):
        if i == 0:
            # Calculate mean and std dev.
            mean_under = np.mean(age_list_under50)
            std_dev_under = np.std(age_list_under50)
            
            mean_over = np.mean(age_list_over50)
            std_dev_over = np.std(age_list_over50)
            
            scores.append((mean_under, std_dev_under, mean_over, std_dev_over))
        
        elif i == 9:
            mean_under = np.mean(hours_list_under50)
            std_dev_under = np.std(hours_list_under50)
            
            mean_over = np.mean(hours_list_over50)
            std_dev_over = np.std(hours_list_over50)
            
            scores.append((mean_under, std_dev_under, mean_over, std_dev_over))
            
        else:
            # Calculate the intersection
            intersection_dictionary = calculateIntersection(train_set, i)

            # Calculates the conditional probabilities
            conditional_dict = calculateConditionalProbability(intersection_dictionary, prior_over50, prior_under50)

            # Add the dictionary with conditional probabilities to scores
            scores.append(conditional_dict)
    
    # print(scores)
    
    return scores, [prior_under50, prior_over50]
```
